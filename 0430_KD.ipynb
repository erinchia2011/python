```python
excelname= "KD"
sheetname = "kd"

wb = openpyxl.load_workbook(str(excelname)+".xlsx")
ws = wb[str(sheetname)]

dataset = []

i = 1
while(ws.cell(row=i, column=1).value!=None):
    row = []
    j = 1
    while(j<=5):#(ws.cell(row=i, column=j).value!=None):
        row.append(ws.cell(row=i, column=j).value)
        j = j +1
    dataset.append(row)
    i = i + 1
```

```python
# trade =[[0 for x in range(11)] for y in range(len(dataset))]
trade = []
for i in range(0, len(dataset)):
    row = []
    for j in range(0, 11):
        row.append(0)
    trade.append(row)
trade[0][0] = "date"
trade[0][1] = "open"
trade[0][2] = "high"
trade[0][3] = "low"
trade[0][4] = "close"
trade[0][5] = "rsv"
trade[0][6] = "k"
trade[0][7] = "d"
trade[0][8] = "buy_day"
trade[0][9] = "cum_day"
trade[0][10] = "sell_day"
```

```python
for i in range(1, len(trade)):   #這邊是把檔案的數據丟進去
    trade[i][0] = dataset[i][0]  #[↓][→]
    trade[i][1] = float(dataset[i][1])
    trade[i][2] = float(dataset[i][2])
    trade[i][3] = float(dataset[i][3])
    trade[i][4] = float(dataset[i][4])    
```

```python
# rsv
for i in range(9,len(trade)):
    rsv=0
    high=[]
    low=[]
    for j in range(0,9):
        high.append(trade[i-j][2])
        low.append(trade[i-j][3])
    Max=max(high)
    Min=min(low)
    rsv=(trade[i][4]-Min)/(Max-Min)*100
    trade[i][5]=rsv
```

```python
#K
trade[9][6]=50
for i in range(10,len(trade)):
    K=0
    K=trade[i-1][6]*2/3+trade[i][5]*1/3
    trade[i][6]=K
    print(K)
```

```python
#D
trade[9][7]=50
for i in range(10,len(trade)):
    D=0
    D=trade[i-1][7]*2/3+trade[i][6]*1/3
    trade[i][7]=D
    print(D)
```

```python
for i in range(10, len(trade)):
    if(trade[i][6] > trade[i][7])and(trade[i-1][8]==0):
        trade[i][8] = 1
        trade[i][9] = 1
        trade[i][10] = 0
    elif(trade[i][6] < trade[i][7])and(trade[i-1][8]==1):
        trade[i][8] = 0
        trade[i][9] = 0
        trade[i][10] = 1
    elif(trade[i-1][2]>0):
        trade[i][8] = 0
        trade[i][9] = 1
        trade[i][10] = 0
    print(trade[i][8],trade[i][9],trade[i][10],"\n")
```
